```python
Question 4:

We fit two models to data with n = 100 observations and one predictor X:
- Linear model: Y = β0 + β1·X + ε
- Cubic model: Y = β0 + β1·X + β2·X² + β3·X³ + ε

(a) Suppose the true relationship between X and Y is linear.
Compare the training RSS for linear vs. cubic regression.

✅ Answer:
The **training RSS for the cubic model will always be less than or equal to** the training RSS for the linear model.

Why?
- The cubic model includes the linear model as a special case (it adds more flexibility).
- More flexibility allows the cubic model to **fit the training data at least as well**, if not better.
- Even if the extra polynomial terms (X², X³) aren't useful, the model can assign their coefficients to zero, reducing to the linear fit.
  
So:  
➡ **Training RSS (Cubic) ≤ Training RSS (Linear)**  
✅ We expect **cubic model to have equal or lower training RSS**.

(b) Suppose the true relationship is linear. Compare **test RSS** for linear vs cubic regression.

✅ Answer:
If the true relationship is linear, then the **linear model is correctly specified**. The cubic model is more flexible, so it risks **overfitting** the noise in the training set.

As a result:
- The linear model will likely generalize **better** to unseen data.
- The **test RSS will generally be lower for the linear model**.

➡ ✅ Expected: **Test RSS (Linear) < Test RSS (Cubic)**

(c) Suppose the true relationship is not linear (but we don't know how far from linear). Compare **training RSS** for linear vs cubic regression.

✅ Answer:
The cubic model is more flexible and can approximate more complex relationships. Therefore, it will again have **lower or equal training RSS** compared to the linear model.

Even if the underlying function is nonlinear, cubic regression can adapt better to curvature.

➡ ✅ Expected: **Training RSS (Cubic) ≤ Training RSS (Linear)**

(d) Same as (c), but for **test RSS**.

✅ Answer:
This is more nuanced. Since the true function is nonlinear but unspecified, we don't know whether the additional flexibility of the cubic model will help or hurt.

- If the nonlinearity is mild: cubic model may **overfit**, leading to higher test RSS.
- If the nonlinearity is strong: cubic model may **capture structure** better, reducing test RSS.

➡ ✅ Conclusion: **Not enough information to tell.**
The test RSS **could be higher or lower** depending on how nonlinear the true relationship is.
```
