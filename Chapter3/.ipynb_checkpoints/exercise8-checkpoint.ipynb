```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
```

```python
from ISLP import load_data
from ISLP.models import (ModelSpec as MS, summarize, poly)
```

```python
import matplotlib.pyplot as plt
```

```python
auto = load_data("Auto")
auto
```

```python
X = pd.DataFrame({'intercept': np.ones(auto.shape[0]),
                  'horsepower': auto['horsepower']})
X = X.reset_index(drop=True)
X
```

```python
y = auto["mpg"].reset_index(drop=True)
y
```

```python
model = sm.OLS(y, X)
results = model.fit()
```

```python
summarize(results)
```

```python
dir(results)
```

```python
new_x = pd.DataFrame({'intercept': [1.0], 'horsepower': [98]})
new_x
```

```python
predicted = results.get_prediction(new_x)
predicted.predicted_mean
```

```python
predicted.predicted
```

```python
dir(predicted)
```

```python
predicted.conf_int(alpha=0.05)
```

```python
predicted.conf_int(alpha=0.05, obs=True)
```

```python
def abline(ax, b, m, *args, **kwargs):
    "Add a line with slope m and intercept b (y = mx + b) to the axis"
    xlim = ax.get_xlim()
    ylim = [m*xlim[0] + b, m*xlim[1] + b]
    ax.plot(xlim, ylim, *args, **kwargs)
```

```python
fig, ax = plt.subplots()
b, m = results.params

ax.scatter(X["horsepower"], y, color='blue', )
abline(ax, b, m, color='green')
ax.set_xlabel("Predictor")
ax.set_ylabel("Response")
```

```python
ax = plt.subplots(figsize =(8 ,8))[1]
ax.scatter(results.fittedvalues, results.resid)
ax.set_xlabel('Fitted value ')
ax.set_ylabel('Residual ')
ax.axhline (0, c='k', ls='--');
```
