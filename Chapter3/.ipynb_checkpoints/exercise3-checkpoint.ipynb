```python
We are given the regression model:

  Salary = 50 + 20·GPA + 0.07·IQ + 35·Level + 0.01·(GPA × IQ) − 10·(GPA × Level)

Where:
- Level = 1 for College, 0 for High School
- GPA and IQ are numeric predictors
- The response is starting salary in thousands

(a) Which answer is correct and why?

To compare college vs high school graduates at fixed GPA and IQ:
- For High School (Level = 0): 
    Salary_HS = 50 + 20·GPA + 0.07·IQ + 0 + 0.01·GPA·IQ + 0
- For College (Level = 1):
    Salary_College = 50 + 20·GPA + 0.07·IQ + 35 + 0.01·GPA·IQ − 10·GPA

Now compute the difference:

    Salary_College − Salary_HS = 35 − 10·GPA

Interpretation:
- If GPA < 3.5 ⇒ College grads earn more
- If GPA > 3.5 ⇒ High school grads earn more

✅ Correct answer: (iii) 
"For a fixed value of IQ and GPA, high school graduates earn more, on average, than college graduates provided that the GPA is high enough."

(b) Predict the salary of a college graduate with IQ = 110 and GPA = 4.0:

Use the equation:

  Salary = 50 + 20·4.0 + 0.07·110 + 35 + 0.01·(4.0 × 110) − 10·4.0
         = 50 + 80 + 7.7 + 35 + 4.4 − 40
         = 137.1

✅ Predicted Salary: $137.1K

(c) True or False: Since the coefficient for the GPA × IQ interaction term is very small, there is very little evidence of an interaction effect.

False.

A small coefficient (0.01) does not by itself imply that the interaction is insignificant. We must look at the **p-value** and **standard error** of the coefficient to determine statistical significance. Also, even small interaction effects can be meaningful depending on the scale of variables involved.

✅ Answer: False — statistical significance requires hypothesis testing, not just visual inspection of the coefficient magnitude.
```
