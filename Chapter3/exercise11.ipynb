```python
import pandas as pd
import numpy as np
```

```python
import statsmodels.api as sm
```

```python
from ISLP import load_data
from ISLP.models import (ModelSpec as MS, summarize, poly)
```

```python
rng = np.random.default_rng(1)
x = rng.normal(size = 100)
y = 2 * x + rng.normal(size = 100)
```

```python
model = sm.OLS(y, x)
results = model.fit()
summarize(results)
```

```python
model2 = sm.OLS(x, y)
results2 = model2.fit()
summarize(results2)
```

(c) We can see that both the part has different coef and standard errors, but t values are equal.
