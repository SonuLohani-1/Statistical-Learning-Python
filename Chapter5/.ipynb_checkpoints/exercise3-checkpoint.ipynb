## (a)

In k-fold cross-validation, we randomly divide the dataset into k folds of approximately equal size. Then we fit k models. In each model we train the model k-1 folds and test the model on the remaining validation set. The prediction error is calculated based on the held out fold.
After repeating this process for all k folds, we average the k validation errors to obtain the cross-validation estimate of the test error.
