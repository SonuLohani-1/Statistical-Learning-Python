```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sample_size = np.arange(1, 10001)
probability = 1- (1-1/sample_size)**sample_size
```

```python
fig, ax = plt.subplots()
ax.plot(sample_size, probability)
ax.set_xlabel("n")
ax.set_ylabel("P(jth obs. in the bootstrap sample)")
ax.set_title("Plots for probability vs n")
```
