```python
import numpy as np
import pandas as pd
```

```python
from ISLP import load_data
from ISLP.models import (ModelSpec as MS, summarize)
from ISLP import confusion_table
```

```python
from sklearn.model_selection import train_test_split
import statsmodels.api as sm
```

# Default Dataset

```python
default = load_data("Default")
default
```

# (a)

```python
design = MS(["balance", "income"])
X = design.fit_transform(default)
y = default["default"] == "Yes"
model = sm.GLM(y, X, family=sm.families.Binomial())
results = model.fit()
```

```python
summarize(results)
```

# (b)

```python
# (i)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
# (ii)
model_b = sm.GLM(y_train, X_train, family=sm.families.Binomial())
results_b = model_b.fit()
summarize(results_b)
```

```python
# (iii)
predictions = results_b.predict(X_test)
labels = np.where(predictions > 0.5, 1, 0)
```

```python
# (iv)
np.sum(labels != y_test)/y_test.shape[0] * 100
```

```python
confusion_table(labels, y_test)
```

# (c)

```python
for random_state in [18, 25, 29]:
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=random_state)
    model = sm.GLM(y_train, X_train, family=sm.families.Binomial())
    results = model.fit()
    print("\n Summary :", "\n", summarize(results), "\n")
    predictions = results.predict(X_test)
    labels = np.where(predictions > 0.5, 1, 0)
    print("validation set error:", np.sum(labels != y_test)/y_test.shape[0] * 100, "%")
    print(confusion_table(labels, y_test))
    
```

# (d)

```python
design = MS(["balance", "income", "student"])
X = design.fit_transform(default)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
model = sm.GLM(y_train, X_train, family=sm.families.Binomial())
results = model.fit()
summarize(results)
```

```python
predictions = results.predict(X_test)
labels = np.where(predictions > 0.5, 1, 0)
np.sum(labels != y_test)/y_test.shape[0] * 100
```

```python
confusion_table(labels, y_test)
```
