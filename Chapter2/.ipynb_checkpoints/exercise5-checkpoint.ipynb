### 5. Solution:

Advantages of a very flexible model against a less flexible model:
(i) If the relationship is not very simple (i.e other than linear model) then a flexible model gives us a better representatin of the data.
(ii) A flexible method capture more of the patterns that are present in the dataset.

Disadvantages of a very flexible model against a less flexible model:
(i) If the relationship is very simple and some variation is there just because of the noise present in the data a flexible will capture those noise in the data as well.
(ii) The bias for a very flexible model is more as compared to a less flexible model (this is because it capture more of the details so some noise gets captured as well).

So, considering this in mind a more flexible model will be preferred when we somehow have an idea data the patterns in the data is not very simple, so a less flexible model won't be able to capture all of the patterns.
On the other hand a less flexible model will be preferred when we expect underlying relation to be simple and the bias to be as minimum as possible.
