### 6. Solution

Differences between a parametric and a non-parametric statistical learning approach:

(i) Parametric approach assumes the underlying function of for the statistical learning while the non-parametric approach doesn't do that.

Advantage of non-Parametric approach: Since this approach doesn't make any approach about the underlying distribution it can be quite useful in figuring out the actual underlying distribution while in the case of parametric approach sometimes the underlying distribution can be very different from the assumption.

Disadvantage of the non-Parametric approach: Here the amount of data required is quite large as compared to the parametric approach.

Some more features better mentioned in a table:

| Aspect            | **Parametric**                | **Non-Parametric**               |
| ----------------- | ----------------------------- | -------------------------------- |
| Assumes function? | ✅ Yes (e.g., linear form)     | ❌ No specific form assumed       |
| Flexibility       | ❌ Less flexible               | ✅ More flexible                  |
| Data requirement  | ✅ Works with small data       | ❌ Needs large datasets           |
| Interpretability  | ✅ Easy to interpret           | ❌ Often harder to interpret      |
| Risk of bias      | ❌ High if model form is wrong | ✅ Low (can fit complex patterns) |
| Risk of variance  | ✅ Lower                       | ❌ Higher (can overfit)           |
